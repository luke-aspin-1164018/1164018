```python
# import IMDB title,crew,names,ratings, datasets, and principals and create dataframes
imdb_title=pd.read_table("dataset/title.basics.tsv",sep='\t')
imdb_crew=pd.read_table("dataset/title.crew.tsv",sep='\t')
imdb_names=pd.read_table("dataset/name.basics.tsv",sep='\t')
imdb_ratings=pd.read_table("dataset/title.ratings.tsv",sep='\t')
# NOTE: principals datasource was too large and required more system resources (especially RAM) than available
# imdb_principals=pd.read_table("dataset/title.principals.tsv",sep='\t')

# Merge titles dataset to crew dataset with left join
imdb_merged = pd.merge(imdb_title,imdb_crew, on='tconst', how='left')
imdb_merged.info()
```

```python
# standardise imdb column data formats (i.e. lower case, title) to equivalent nds dataframe columns formats to avoid match errors later 
imdb_merged['tconst']=imdb_merged['tconst'].str.strip().str.lower()
imdb_merged['primaryTitle']=imdb_merged['primaryTitle'].str.strip().str.title()
imdb_merged['originalTitle']=imdb_merged['originalTitle'].str.strip().str.title() 
imdb_merged['directors']=imdb_merged['directors'].str.strip().str.lower()
imdb_names['nconst']=imdb_names['nconst'].str.strip().str.lower()
imdb_names['primaryName']=imdb_names['primaryName'].str.strip().str.title()
imdb_ratings['tconst']=imdb_ratings['tconst'].str.strip().str.lower()
# NOTE: principals datasource was too large and required more system resources (especially RAM) than available
# imdb_principals['job']=imdb_principals['job'].str.strip().str.lower()
# imdb_principals['nconst']=imdb_principals['nconst'].str.strip().str.lower()
```

```python

# iterate through nds dataframe rows with missing Director column values, and compare runtime/title/release year values to equivalents in IMDB dataframe, then get nconst (IMDB director ID) and insert into nds dataframe Director column 
for n in nds[(nds['Director'].isnull())].itertuples():
    # query IMDB datasets for Directors name(s) based on match with movie runtime (+/- 2 mins), NDS title matching the IMDB primary OR original title, and with IMDB 'startYear' (or earlier) matching the year it showed in Netflix
    nconsts = imdb_merged.loc[((imdb_merged['runtimeMinutes'] >= str(n[7]-2)) & (imdb_merged['runtimeMinutes'] <= str(n[7]+2))) & ((imdb_merged['primaryTitle']==n[1]) | (imdb_merged['originalTitle']==n[1])) & (imdb_merged['startYear'] <= str(n[5].year)), 'directors']
    imdb_director =""
    # if the Director entry in IMDB DS is not NULL, ie starts with 'nm'
    if str(nconsts).split()[1].startswith("nm"): 
        nconsts=str(nconsts).split()[1] # get the IMDB DS value of Directors
        if ',' in nconsts: # if the IMDB DS Directors value includes commas (ie multiple directors/nconst values), loop through value and lookup each primaryName that corresponds to each nconst
            for index, nconst in enumerate(nconsts.split(',')):
                imdb_director_lookup = imdb_names.loc[imdb_names["nconst"] == nconst, "primaryName" ].item()
                #store all director names as comma separated string
                imdb_director += imdb_director_lookup if index==nconsts.count(',') else f"{imdb_director_lookup},"
        else: # if IMDB DS Directors contains only one nconst value, lookup the primaryName 
            imdb_director = imdb_names.loc[imdb_names["nconst"] == nconsts, "primaryName" ].item()
        nds.loc[n[0], 'Director'] = imdb_director # assign the directors string to the nds Director field for the relevant row
```

```python
nds.info()
```

```python
# NOTE: this takes a very long time to process -- approx 4 hours
# iterate through all nds dataframe rows and compare runtime/title/release year to equivalents in IMDB dataframe, then get tconst (IMDB title ID) and insert into new nds dataframe column 'imdb_tt'
for n in nds.itertuples():
    # query IMDB datasets for tconst (IMDB title ID) based on match with movie runtime (+/- 2 mins), NDS title matching the IMDB primary OR original title, and with IMDB 'startYear' (or earlier) matching the year it showed in Netflix
    tconst = imdb_merged.loc[((imdb_merged['runtimeMinutes'] >= str(n[7]-2)) & (imdb_merged['runtimeMinutes'] <= str(n[7]+2))) & ((imdb_merged['primaryTitle']==n[1]) | (imdb_merged['originalTitle']==n[1])) & (imdb_merged['startYear'] <= str(n[5].year)), 'tconst']
    if str(tconst).split()[1].startswith("tt"):
        nds.loc[n[0], 'imdb_tt'] = str(tconst).split()[1] # assign the tconst (IMDB title ID) string to the nds dataframe imdb_tt column for the relevant row
```

```python
#iterate through all nds dataframe rows and use imdb_tt (IMDB tconst title ID) to get imdb_ratings dataframe averageRating and numVotes values and insert into new columns in nds dataframe  

# function to determine if string is float
def isFloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False
 
for n in nds.itertuples():
    # check that row from nds dataframe contains tconst string
    if str(n[10]).startswith("tt"): 
        # collecft averageRating and numVotes values for tconst string from title.ratings.tsv
        rating = imdb_ratings.loc[(imdb_ratings['tconst'] == n[10]), ['averageRating', 'numVotes']]
        averageRating=str(rating['averageRating']).split()[1]
        numVotes=str(rating['numVotes']).split()[1]
        # check tt number from nds dataframe has match in title.ratings.tsv by verifying if returned rating is float
        if isFloat(averageRating):
            nds.loc[n[0], 'IMDB_Rating'] = float(averageRating) #insert rating value into new column
            nds.loc[n[0], 'IMDB_Votes'] = int(numVotes) #insert votes value into new column
nds.info()
```

```python
# create lists to store index values of missing IMDB_Rating, IMDB_Votes, Cast values

# first reset index as it is unclean with entries populated across non-contiguous index values
nds.reset_index(inplace=True)

missing_IMDB_Rating = nds[nds["IMDB_Rating"].isnull()].index.to_list()
missing_IMDB_Votes = nds[nds["IMDB_Votes"].isnull()].index.to_list()
missing_Cast = nds[nds["Cast"].isnull()].index.to_list()
```

```python
# Impute median values for all nds dataframe rows with missing IMDB_Rating values
nds.loc[nds['IMDB_Rating'].isna(), 'IMDB_Rating'] = nds['IMDB_Rating'].median()
# check first few updated (previously missing) IMDB_Rating values
nds.iloc[missing_IMDB_Rating].head(10)
```

```python
# Impute median values for all nds dataframe rows with missing IMDB_Votes values
nds.loc[nds['IMDB_Votes'].isna(), 'IMDB_Votes'] = int(nds['IMDB_Votes'].median())
# check first few updated (previously missing) IMDB_Votes values
nds.iloc[missing_IMDB_Votes].head(10)
```

```python
# For all missing cast values, fill with 'NA' - some titles have no cast (eg just a recoding of a specific scene)
nds.loc[nds['Cast'].isna(), 'Cast'] = "NA"
# check first few updated (previously missing) Cast values
nds.iloc[missing_Cast].head(10)
```

```python
# drop imdb_tt column as now irrelevant
nds = nds.drop('imdb_tt', axis=1)
# convert IMDB_Votes back to integer - was automatically converted to float with above impute of missing values
nds['IMDB_Votes'] = pd.to_numeric(nds['IMDB_Votes'], downcast='integer', errors='coerce')
nds.info()
```

```python
# Display a count of outliers for each numerical feature using z-score and modified z-score methods

nds_z_scores = nds.copy()

# function to calculate z-scores
def z_score(feature):
    # calculate z-score value
    nds_z_scores['Z_'+ feature] = np.abs(stats.zscore(nds_z_scores[feature]))
    return nds_z_scores

#function to calculate modified z-scores
def mod_z_score(feature):
   #calculate median of MAD ( the median absolute deviation value) 
   mad = (np.abs(nds_z_scores[feature] - nds_z_scores[feature].median())).median()
   # calculate the modified z-score value
   nds_z_scores['Z_'+  feature] = 0.6745 * (nds_z_scores[feature] - nds_z_scores[feature].median()) / mad
   return nds_z_scores

nds_z_scores_cleaned = nds.copy()

# calculate and print a count of z scores greater than 3 for Duration and IMDB_Rating. 
for f in ['Duration','IMDB_Rating']:
    nds_z_scores = z_score(f)
    print(f"Feature {f} outliers (Z-Score): {len(nds_z_scores[nds_z_scores['Z_'+ f] >= 3])}")
    

# calculate and print a count of modified z scores greater than 3.5 for Release_Year and IMDB_Votes. 
features = ['Release_Year','IMDB_Votes']                                
for f in features:
    nds_z_scores = mod_z_score(f)
    print(f"Feature {f} outliers (Modified Z-Score): {len(nds_z_scores[nds_z_scores['Z_'+ f] >= 3.5])}")
```

```python
# Quantify and visualise the outliers with scatter plots - how extreme are they?
# NOTE: Release_Year has 0 outliers, so no scatter plot for that feature 

features = ['Duration','IMDB_Rating','IMDB_Votes']

plt.figure(figsize=(16,16))
for f in features:
    score = 3.5 if f == 'IMDB_Votes' else 3
    outliers = nds_z_scores[nds_z_scores['Z_'+ f].abs() > score]
    plt.figure(figsize=(10, 6))
    plt.scatter(nds_z_scores[f].index, nds_z_scores['Z_'+ f], label='Data Points', color='yellowgreen')
    plt.scatter(outliers[f].index, outliers['Z_'+ f], color='red', label='Outlier')
    plt.xlabel('Index Value')
    plt.ylabel('Z-score')
    plt.title(f' Feature {f} vs. Z-score  - Scatter Plot (Total Outliers: {len(outliers)})')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Create new dataframe without outliers
nds_z_scores_cleaned = nds_z_scores[(nds_z_scores.Z_Duration < 3) &(nds_z_scores.Z_Duration > -3) & (nds_z_scores.Z_IMDB_Rating < 3) & (nds_z_scores.Z_IMDB_Rating > -3) & (nds_z_scores.Z_Release_Year < 3.5) & (nds_z_scores.Z_Release_Year > -3.5) & (nds_z_scores.Z_IMDB_Votes < 3.5) & (nds_z_scores.Z_IMDB_Votes > -3.5) ] 

# display original dataframe to cleaned dataframe mean and median (rounded to nearest integer)
print("*" * 20)
for f in features:
    print(f"{f} original Mean: {round(nds_z_scores[f].mean())} ; Cleaned Mean: {round(nds_z_scores_cleaned[f].mean())}")
    print(f"{f} original Median: {round(nds_z_scores[f].median())} ; Cleaned Median: {round(nds_z_scores_cleaned[f].median())}")
    print("*" * 20)
```

```python
# compare feature data with and without outliers using probability plots - does removing outlier data points significantly change the STD, mean/median, or best fit line?
for f in features:
    sns.set_style('whitegrid')
    sns.despine()
    plt.figure(figsize=(16,6))

    plt.subplot(1, 2, 1)
    stats.probplot(nds_z_scores[f], plot=plt)

    plt.subplot(1, 2, 2)
    stats.probplot(nds_z_scores_cleaned[f], plot=plt)
```

```python
# pie charts shows Top 10 Directors/Actors/Countries output as a % of total activity by all Directors/Actors/Countries (not total titles) done in dataframe
# NOTE: titles can have multiple Directors/Actors/Countries

# sum the number of titles worked on by Top 10 for each category
directors_t10_total = directors.sum()
actors_t10_total = actors.sum()
countries_t10_total = countries.sum()

# sum the total number of titles worked on by all from each category
directors_total = pd.Series([a for inner in nds_eda['Director'].dropna().str.split(', ').tolist() for a in inner]).value_counts().sum()
actors_total = pd.Series([a for inner in nds_eda['Cast'].dropna().str.split(', ').tolist() for a in inner]).value_counts().sum()
countries_total = pd.Series([c for inner in nds_eda['Country'].dropna().str.split(', ').tolist() for c in inner]).value_counts().sum()

# calculate percentages for T10 activity and total activity without Top 10 activity
d_list = [directors_t10_total,directors_total - directors_t10_total]
a_list = [actors_t10_total,actors_total - actors_t10_total]
c_list = [countries_t10_total,countries_total - countries_t10_total]
labels = ["Top10","Remainder"]

plt.figure(figsize=(16, 6))
plt.subplot(1, 3, 1)
plt.pie(d_list, labels=labels, startangle=90, autopct='%1.1f%%')
plt.title('Top 10 Directors\' as a % of total activity')

plt.subplot(1, 3, 2)
plt.pie(a_list, labels=labels, startangle=90, autopct='%1.1f%%')
plt.title('Top 10 Actors\' as a % of total activity')

plt.subplot(1, 3, 3)
plt.pie(c_list, labels=labels, startangle=90, autopct='%1.1f%%')
plt.title('Top 10 Countries\' as a % of total activity')

plt.show()
```

___
### Insights
- Most movies are directed by unique director/s, with the Top 10 Directors responsible for only 3% of all Director activity in the dataframe
- Most movies have unique actors, with the Top 10 actors responsible for only 0.6% of the total work done by all actors in the dataframe
- The Top 10 countries are responsible for 77.6% of all activity in the dataframe
- The Top 10 actors come from the top 3 countries: USA, India, UK
- Age ratings skew to more mature audiences - which is understandable given Netflix is a subscription service. Most of their content will be aimed at the bill payers.
